# Weighted Voronoi
This notebook calculates the weighted Voronoi matrix and shows how to plot it. It is based on the script `voronoi_weighted.py`.

```python
from voronoi_weighted import weighted_voronoi,plot_voronoi
%matplotlib inline
```

Give player positions

```python
home_players_positions = [(10,30),(35,21),(34,30),(36,37),(38,39),(40,40),(40,32),(55,35),(56,50),(55,58),(60,50)]
away_players_positions = [(34,20),(37,34),(37,40),(40,50),(55,15),(54,32),(56,45),(58,56),(70,28),(72,45),(80,35)]
```

Calculate weighted Voronoi matrix and plot

```python
wv = weighted_voronoi(home_players_positions,away_players_positions,resolution=200, beta=2)
plot_voronoi(wv, figsize=(20,12), lc = 'black', interpolation = 'none', cmap = 'seismic', scale = 1, axis = 'off')
```
